```python
from datasets import load_dataset, DownloadConfig
import pandas as pd

ds2 = load_dataset("drveronika/x_fake_profile_detection", split="train", streaming=True, features=None)
records = []
for i, example in enumerate(ds2):
    records.append({k: str(v) for k, v in example.items() if k != "image"})
    if i >= 9999: break

df2 = pd.DataFrame(records)
print(df2.columns.tolist())
df2.to_csv("x_fake_profile_detection.csv", index=False)


```
